```python
# -*- coding: utf-8 -*-
import cv2,math,glob
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
#plt.figure(figsize=(16,16))

#lv_source_path = '/Users/hebing/working/2018/Invoice_recognition/Data/invoice/Invoice_split_blue/debug'
lv_source_path = 'image/origin'
lv_target_path = 'image/output'

##Define threshold for blue words
#[157  90  66]
lower_blue=np.array([78,90,67])
upper_blue=np.array([157,200,200])

# lower_blue=np.array([78,43,46])
# upper_blue=np.array([110,255,255])

# lower_blue=np.array([110,50,50])
# upper_blue=np.array([130,255,255])

print_flag=0
print_result_flag=1

if os.path.exists(lv_target_path) == False:
    shutil.rmtree(lv_target_path)
    os.makedirs(lv_target_path)  
    
lv_index=0
lv_filenames = glob.glob(os.path.join(lv_source_path, '*.jpg'))
for lv_file in lv_filenames[0:1]:
    #f = img_dir+'PP0002-002fe69b98be214cc2ce81aba7dd4fb9.jpg'
    print(lv_index,len(lv_filenames),lv_file)
    lv_index = lv_index + 1
    lv_basename = os.path.basename(lv_file)
    lv_target_filename = os.path.join(lv_target_path, lv_basename + '_blue.jpg')
    im_original = cv2.imread(lv_file)

    hsv = cv2.cvtColor(im_original, cv2.COLOR_BGR2HSV)
    mask = cv2.inRange(hsv, lower_blue,upper_blue)
    output = cv2.bitwise_and(im_original, im_original, mask=mask)
    cv2.imwrite("t1.jpg", output)

    if print_flag == 1:
        plt.figure(figsize=(16,16))
        plt.imshow(output[:,:,::-1])
        plt.show()

    b,g,r = im_original[:,:,0],im_original[:,:,1],im_original[:,:,2]
    mean = np.mean(im_original)
    mean_b,mean_g,mean_r = np.mean(b),np.mean(g),np.mean(r)
    #print 'Mean is:',mean
    #black[black>mean] = 0
    output[output[:,:,1]>mean_g] = 0
    output[output[:,:,2]>mean_r] = 0
    if print_flag == 1:
        plt.figure(figsize=(16,16))
        plt.imshow(output[:,:,(2,1,0)])
        plt.show()
    cv2.imwrite("t2.jpg", output)
    gray = cv2.cvtColor(output, cv2.COLOR_BGR2GRAY)
    gradX = cv2.Sobel(gray, ddepth = cv2.CV_32F, dx = 1, dy = 0, ksize = -1)
    gradY = cv2.Sobel(gray, ddepth = cv2.CV_32F, dx = 0, dy = 1, ksize = -1)
    gradient = cv2.subtract(gradX, gradY)
    gradient = cv2.convertScaleAbs(gradient)
    if print_flag == 1:
        plt.figure(figsize=(16,16))
        plt.imshow(gradient,'gray')
        plt.show()

    blurred = cv2.blur(gradient, (9, 9))
    (_, thresh) = cv2.threshold(blurred, 225, 255, cv2.THRESH_BINARY)
    if print_flag == 1:
        plt.figure(figsize=(16,16))
        plt.imshow(blurred,'gray')
        plt.show()

    closed = cv2.erode(blurred, None, iterations = 6)
    blurred = cv2.dilate(blurred, None, iterations = 6)
    if print_flag == 1:
        plt.figure(figsize=(16,16))
        plt.imshow(blurred,'gray')
        plt.show()

    _, cnts, _= cv2.findContours(blurred.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    img_h,img_w,img_c = im_original.shape
    min_area,max_area = 0,img_w/2.0*img_h/2
    boxes = []
    print(len(cnts))
    for (i, c) in enumerate(cnts):
        (x, y, w, h) = cv2.boundingRect(c)
        if w>0.2*h:
            if w*h>min_area and w*h<max_area: #scale and area filter
                if x>=50 and x <=img_w-100: #position filter
                    if y>=50 and y <=img_h-50:
                        boxes.append((x,y,w,h))
                        #cv2.rectangle(im_original,(x,y),(x+w,y+h),(0,0,255),2)
                        
    result = 255*np.ones((img_h,img_w,img_c),im_original.dtype)
    for x,y,w,h in boxes:
        result[y:y+h,x:x+w,:] = im_original[y:y+h,x:x+w,:]
    if print_result_flag == 1:
        plt.figure(figsize=(16,16))
        plt.imshow(result[:,:,::-1])
        plt.show()

    cv2.imwrite(lv_target_filename,result)

print('Done')
```
